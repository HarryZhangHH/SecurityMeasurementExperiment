transformation

```python
def transform(data):
    data = data.loc[:, ['Which Group', 'Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']]
    data['Which Group'] = data['Which Group'].str.extract('(\d+)').astype(int)
    difficulty_map = {
        '0-20%': 1,
        '20-40%': 2,
        '40-60%': 3,
        '60-80%': 4,
        '80%-100%': 5
    }
    data[['Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']] = data[['Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']].replace(difficulty_map)
    return data
```

## perspective data structure

the data structure storing score for each group

```python
new_score = {
    'dos_f': [],
    'dos_s': [],
    'path_f': [],
    'path_s': [],
    'user_f': [],
    'user_s': [],
    'xss_f': [],
    'xss_s': []
}

old_score = {
    'dos_f': [],
    'dos_s': [],
    'path_f': [],
    'path_s': [],
    'user_f': [],
    'user_s': [],
    'xss_f': [],
    'xss_s': []
}

complete_score = {
    'dos_f': [],
    'dos_s': [],
    'path_f': [],
    'path_s': [],
    'user_f': [],
    'user_s': [],
    'xss_f': [],
    'xss_s': []
}

group_to_files = {
    1: ['dos_s', 'path_f', 'user_s', 'xss_f'],
    2: ['dos_f', 'path_s', 'user_f', 'xss_s'],
    3: ['dos_f', 'path_s', 'user_s', 'xss_f'],
    4: ['dos_s', 'path_f', 'user_f', 'xss_s'],
    5: ['dos_f', 'path_f', 'user_s', 'xss_s'],
    6: ['dos_s', 'path_s', 'user_f', 'xss_f']
}

def add_new_score(file_name = 'dos_f', value = 1):
    new_score[file_name].append(value)

def add_old_score(file_name = 'dos_f', value = 1):
    old_score[file_name].append(value)

def add_complete_score(file_name = 'dos_f', value = 1):
    complete_score[file_name].append(value)

def get_new_score(file_name = 'dos_f'):
    return new_score[file_name]

def get_old_score(file_name = 'dos_f'):
    return old_score[file_name]

def get_complete_score(file_name = 'dos_f'):
    return complete_score[file_name]
```

change the original form structure

```python
def fill_new_score(row):
    for i in range(4):
        add_new_score(file_name=group_to_files[row['Which Group']][i], value=row['Difficulty.Correct_' + str(i + 1)])

def fill_old_score(row):
    for i in range(4):
        add_old_score(file_name=group_to_files[row['Which Group']][i], value=row['Difficulty.Correct_' + str(i + 1)])

def fill_complete_score(row):
    for i in range(4):
        add_complete_score(file_name=group_to_files[row['Which Group']][i], value=row['Difficulty.Correct_' + str(i + 1)])
```

```python
new_data.apply(fill_new_score, axis=1)
new_score
old_data.apply(fill_old_score, axis=1)
old_score
complete_data.apply(fill_complete_score, axis=1)
complete_score
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

def plot(score = new_score):
    for key in score:
        plt.boxplot(score[key])
        plt.ylabel('Usefulness')
        plt.title(key)
        plt.show()

    for key in score:
        sm.qqplot(np.array(score[key]), line='s')
        plt.title(key)
        plt.show()
```

```python
from scipy.stats import mannwhitneyu

def print_statistics(get_score = get_new_score):
    statistic, pvalue = mannwhitneyu(get_score('dos_f'), get_score('dos_s'), alternative='greater')
    mean = np.mean(get_score('dos_f'))
    stdev = np.std(get_score('dos_f'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    mean = np.mean(get_score('dos_s'))
    stdev = np.std(get_score('dos_s'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    print("Mann-Whitney U statistic:", statistic)
    print("p-value:", pvalue)

    statistic, pvalue = mannwhitneyu(get_score('path_f'), get_score('path_s'), alternative='greater')
    mean = np.mean(get_score('path_f'))
    stdev = np.std(get_score('path_f'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    mean = np.mean(get_score('path_s'))
    stdev = np.std(get_score('path_s'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    print("Mann-Whitney U statistic:", statistic)
    print("p-value:", pvalue)

    statistic, pvalue = mannwhitneyu(get_score('user_f'), get_score('user_s'), alternative='greater')
    mean = np.mean(get_score('user_f'))
    stdev = np.std(get_score('user_f'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    mean = np.mean(get_score('user_s'))
    stdev = np.std(get_score('user_s'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    print("Mann-Whitney U statistic:", statistic)
    print("p-value:", pvalue)

    statistic, pvalue = mannwhitneyu(get_score('xss_f'), get_score('xss_s'), alternative='greater')
    mean = np.mean(get_score('xss_f'))
    stdev = np.std(get_score('xss_f'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    mean = np.mean(get_score('xss_s'))
    stdev = np.std(get_score('xss_s'))
    print("Mean of list:", mean)
    print("Standard deviation of list:", stdev)
    print("Mann-Whitney U statistic:", statistic)
    print("p-value:", pvalue)
```
