transformation

```python
def transform(data):
    data = data.loc[:, ['Which Group', 'Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']]
    data['Which Group'] = data['Which Group'].str.extract('(\d+)').astype(int)
    difficulty_map = {
        '0-20%': 1,
        '20-40%': 2,
        '40-60%': 3,
        '60-80%': 4,
        '80%-100%': 5
    }
    data[['Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']] = data[['Difficulty.Correct_1', 'Difficulty.Correct_2', 'Difficulty.Correct_3', 'Difficulty.Correct_4']].replace(difficulty_map)
    return data
```

## perspective data structure

the data structure storing score for each group

```python
score = {
    'dos_f': [],
    'dos_s': [],
    'path_f': [],
    'path_s': [],
    'user_f': [],
    'user_s': [],
    'xss_f': [],
    'xss_s': []
}

group_to_files = {
    1: ['dos_s', 'path_f', 'user_s', 'xss_f'],
    2: ['dos_f', 'path_s', 'user_f', 'xss_s'],
    3: ['dos_f', 'path_s', 'user_s', 'xss_f'],
    4: ['dos_s', 'path_f', 'user_f', 'xss_s'],
    5: ['dos_f', 'path_f', 'user_s', 'xss_s'],
    6: ['dos_s', 'path_s', 'user_f', 'xss_f']
}

def add_score(file_name = 'dos_f', value = 1):
    score[file_name].append(value)

def get_score(file_name = 'dos_f'):
    return score[file_name]
```

change the original form structure

```python
def fill_score(row):
    for i in range(4):
        add_score(file_name=group_to_files[row['Which Group']][i], value=row['Difficulty.Correct_' + str(i + 1)])
```

```python
complete_data.apply(fill_score, axis=1)
score
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

for key in score:
    plt.boxplot(score[key])
    plt.title(key)
    plt.show()

for key in score:
    sm.qqplot(np.array(score[key]), line='s')
    plt.title(key)
    plt.show()
```

```python
from scipy.stats import mannwhitneyu

statistic, pvalue = mannwhitneyu(get_score('dos_f'), get_score('dos_s'))
mean = np.mean(get_score('dos_f'))
stdev = np.std(get_score('dos_f'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
mean = np.mean(get_score('dos_s'))
stdev = np.std(get_score('dos_s'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
print("Mann-Whitney U statistic:", statistic)
print("p-value:", pvalue)

statistic, pvalue = mannwhitneyu(get_score('path_f'), get_score('path_s'))
mean = np.mean(get_score('path_f'))
stdev = np.std(get_score('path_f'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
mean = np.mean(get_score('path_s'))
stdev = np.std(get_score('path_s'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
print("Mann-Whitney U statistic:", statistic)
print("p-value:", pvalue)

statistic, pvalue = mannwhitneyu(get_score('user_f'), get_score('user_s'))
mean = np.mean(get_score('user_f'))
stdev = np.std(get_score('user_f'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
mean = np.mean(get_score('user_s'))
stdev = np.std(get_score('user_s'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
print("Mann-Whitney U statistic:", statistic)
print("p-value:", pvalue)

statistic, pvalue = mannwhitneyu(get_score('xss_f'), get_score('xss_s'))
mean = np.mean(get_score('xss_f'))
stdev = np.std(get_score('xss_f'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
mean = np.mean(get_score('xss_s'))
stdev = np.std(get_score('xss_s'))
print("Mean of list:", mean)
print("Standard deviation of list:", stdev)
print("Mann-Whitney U statistic:", statistic)
print("p-value:", pvalue)
```
